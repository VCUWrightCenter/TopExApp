```python
import medtop.core as medtop
```

```python
data, doc_df = medtop.import_from_files('../../data/corpus_file_list.txt')
data.shape
```

```python
clusteringMethod = 'hac'
k=None
height=None

tfidf, dictionary = medtop.create_tfidf(doc_df,seed_topics_df=None)
data = medtop.get_phrases(data, dictionary.token2id, tfidf, include_input_in_tfidf = True, window_size=6, include_sentiment=True)
data = medtop.get_vectors("tfidf", data, dictionary = dictionary, tfidf = tfidf)
```

```python
medtop.assign_clusters(data, method='hac', k=None, height=None, dist_metric='euclidean')
```
